```python
#plt.subplot(no of  rows,no of columns , indexes)

import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[10,30,20,40]
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title=("Line Chart")

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title=("Bar Chart")

#plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt


#fig , ax = plt.subplots(nrows,ncols,figsize(width,height))
fig,ax= plt.subplots(1,2,figsize=(10,5))
x=[1,2,3,4]
y=[10,30,20,40]

ax[0].plot(x,y, color='orange')
ax[0].set_title("Line plot")

ax[1].bar(x,y,color="red")
ax[1].set_title('Bar Chart')
fig.suptitle("Comparison of two charts line and bar chart")

plt.tight_layout()
plt.show()
```
