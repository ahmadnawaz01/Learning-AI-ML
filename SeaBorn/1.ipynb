```python
import numpy as np
import pandas as pd
import seaborn as sns


```

```python
sns.get_dataset_names()
```

```python
species=sns.load_dataset('penguins')
species.head()
```

```python
species['species'].value_counts()
```

```python
species['island'].value_counts()    
```

```python
sns.set_style('ticks')
sns.set_context('notebook')
sns.scatterplot(data = species, x='flipper_length_mm',y='body_mass_g', hue='sex' , palette='Set1')
sns.despine(left=True)
```

```python
sns.scatterplot(data=species, x='species', y='body_mass_g', hue='island'
,style='sex')
```

```python
sns.stripplot(data=species, x='species', y='body_mass_g', hue='island',dodge=True, jitter=True )
```

```python
sns.set_context('talk')
sns.swarmplot(data=species,x='species',y='body_mass_g', hue='island')
sns.despine()
```

```python
sns.histplot(data=species, x='body_mass_g' , hue='sex', multiple='stack')
```

```python
sns.regplot(data=species, x='body_mass_g', y='flipper_length_mm' , color='yellow')
```

```python
sns.set_style('whitegrid')
sns.lineplot(data=species, x='body_mass_g', y='flipper_length_mm' , hue='sex')
```

```python
sns.jointplot(data=species, x='body_mass_g', y='flipper_length_mm' , hue='sex')
```

```python
sns.jointplot(data=species, x='body_mass_g', y='flipper_length_mm',hue='sex' ,kind='kde')
```

```python
sns.barplot(data=species,x='species',y='body_mass_g', hue='sex' ,palette=['gold','red'], estimator=np.var)
```

```python
sns.countplot(data=species, x='species', hue='island')
```

```python
sns.boxplot(data=species, x='species', y='body_mass_g', hue='sex')
```

```python
sns.violinplot(data=species, x='species', y='body_mass_g',hue='sex', split=True)
```

```python
sns.violinplot(data=species, x='species', y='body_mass_g')
sns.swarmplot(data=species,x='species', y='body_mass_g',hue='island', size=3)
```

```python
sns.kdeplot(data=species, x='body_mass_g' ,hue='species' , fill=True )
```

```python
col=['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']
sns.heatmap(data=species[col].corr(), annot=True, cmap='Blues' ,vmin=0.2)
```

```python
sns.rugplot(data=species, x='body_mass_g' ,height=0.2, hue='species')
```

```python
sns.pairplot(data=species, hue='sex', palette='Set2', diag_kind='hist')
```

```python
graph=sns.PairGrid(data=species, hue='sex', palette='Set2')
graph.map_upper(sns.scatterplot)
graph.map_lower(sns.kdeplot)
graph.map_diag(sns.histplot)
```
