```python
def funstr(st,n):
    for i in range(n):
        print(st," ")

funstr("ahmad",6)
```

```python
class message:
    def __init__(self,n):
        self.message=n
    def displayinformation(self):
        print("message is =",self.message)
        print("reverse message is =",self.message[::-1])
        print("length is =",len(self.message))


my= message("ahmadnawaz")
my.displayinformation()
```

```python

def tempfun(l,val):
    for i in range(len(l)):
        val=val+5
        l[i]=val
l=[25,15,4,2,55,23,32]
tempfun(l,15)

```

```python
def compare(l):
    for i in range(len(l)):
        if l[i]>32:
            print("temperature is hot today")
        else:
            print("temperature is normal today")
        
l=[25,15,44,2,55,23,32]
compare(l)
```
