```python
from collections import deque
```

```python
def bfs(g, startnode,goalnode):
    queue=deque([startnode])
    visited={startnode}
    parent={startnode:None}

    while queue:
        currnode=queue.popleft()
        if(currnode==goalnode):
            path=[]
            node=goalnode
            while node is not None:
                path.append(node)
                node=parent[node]
            path.reverse()
            return path

        for neighbor in graph.get(currnode,[]):
            if neighbor not in visited:
                visited.add(neighbor)
                parent[neighbor]=currnode
                queue.append(neighbor)

    return none



bfs(graph, 'A', 'F')
```

```python
import heapq
g2={
    'A':[("B",1),("C",4)],
    'B':[('D',2)],
    'C':[('D',2)],
    'D':[]
}

def ucs(graph, startnode, goalnode):
    front=[(0,startnode)]
    parent={startnode:None}
    bestcost={startnode:0}
    while front:
        currcost,currnode=heapq.heappop(front)

        if currcost!= bestcost[currnode]:
            continue

        if currnode==goalnode:
            path=[]
            node=goalnode
            while node is not None:
                path.append(node)
                node=parent[node]
            path.reverse()
            return path,currcost


        for ne,edgecost in graph.get(currnode,[]):
            newcost=currcost+edgecost
            if ne not in bestcost or newcost < bestcost[ne]:
                 bestcost[ne]=newcost
                 parent[ne]=currnode
                 heapq.heappush(front,(newcost, ne))
    return None,"0"

ucs(g2,'A','D')
```

```python
g2 = {"A": [("B", 1), ("C", 4)], "B": [("D", 2)], "C": [("D", 1)], "D": []}
h  = {"A": 3, "B": 2, "C": 1, "D": 0}

def astar(g,startnode,goalnode,h):
    frontier=[(h[startnode],0,startnode)]
    parent={startnode:None}
    bestcost={startnode:0}

    while frontier:
        currf, currg,currnode=heapq.heappop(frontier)
        if(currg != bestcost[currnode]):
            continue

        if currnode==goalnode:
            path=[]
            node=goalnode
            while node is not None:
                path.append(node)
                node=parent[node]
            path.reverse()
            return path,currg

        for ne,edgecost in g.get(currnode,[]):
            newg=currg+edgecost
            if ne not in bestcost or newg < bestcost[ne]:
                bestcost[ne]=newg
                parent[ne]=currnode
                newf= newg+h[ne]
                heapq.heappush(frontier,(newf,newg,ne))
    return None, 0

astar(g2,"A","D",h)
```
