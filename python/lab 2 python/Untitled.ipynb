```python
from collections import deque

graph = {
    'S': [('A', 3), ('C', 2), ('D', 2)],
    'A': [],
    'B': [('E', 2)],
    'C': [('F', 1)],
    'D': [('B', 3), ('G', 8)],
    'E': [('G', 2)],
    'F': [('E', 0), ('G', 4)],
    'G': []
}

def BFS(graph, start, goal):
    queue = deque()
    queue.append((start, [start], 0))   # node, path, cost
    explored = []

    visitedal = set()
    parent = {}
    cost = {}

    visitedal.add(start)
    parent[start] = None
    cost[start] = 0

    while queue:
        node, path, total_cost = queue.popleft()

        if node not in explored:
            explored.append(node)

        if node == goal:
            return explored, path, total_cost

        tie_breaker = sorted(graph[node], key=lambda x: x[0])

        for child, edge_cost in tie_breaker:
            if child not in visitedal:
                visitedal.add(child)
                parent[child] = node
                cost[child] = total_cost + edge_cost
                queue.append((child, path + [child], total_cost + edge_cost))

    return explored, [], 0


startnode = 'S'
goalnode = 'G'

explored, path, total_cost = BFS(graph, startnode, goalnode)

print("Explored nodes:", explored)
print("Path:", path)
print("Total cost:", total_cost)
```
