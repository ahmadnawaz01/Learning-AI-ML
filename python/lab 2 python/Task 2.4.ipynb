```python
def findsum(num):
    num.sort() 
    total = sum(num)
    m = total / len(num)
    n = len(num)
    
    if n % 2 == 0:
        median = (num[n//2 - 1] + num[n//2]) / 2 
    else:
        median = num[n//2]

    print("Sum is =", total)
    print("Mean is =", m)
    print("Median is =", median)

numbers = list(map(int, input("Enter the numbers: ").split()))

findsum(numbers)
```
