```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr1=np.array([[1,2,3,4],[4,5,56,7]])
arr2=np.random.rand(3,3)
arr3=np.zeros((3,4))

np.save("arr1.npy",arr1)
np.save("arr2.npy",arr2)
np.save("arr3.npy",arr3)

loaded_arr=np.load("arr1.npy")
print("loaded_arr = ",loaded_arr)
```

```python
numpylogo=np.load("numpylogo.npy")
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.imshow(numpylogo)
plt.title("NumPy logo")
plt.grid(False)


darklogo=1-numpylogo
plt.subplot(122)
plt.imshow(darklogo)
plt.title("NumPy logo")
plt.grid(False)


```
