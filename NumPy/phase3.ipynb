```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("==== Sales Analysis ======")
print("Shape of data = ",sales_data.shape)
print("Sales Data of first three restarnants =\n",sales_data[:3])
```

```python
yearly_sum =np.sum(sales_data[:,1:],axis=1)
print("yealy sum =",yearly_sum)

print("yearly sum =",np.sum(sales_data[:,1:],axis=0))
```

```python
#minimum sales per restaurant

min_sales=np.min(sales_data[:,1:],axis=1)
print("Minium sales per restaurant are = ",min_sales)


##maximum sales per year
max_year=np.max(sales_data[:,1:],axis=0)
print("Maximum Sales per year =",max_year)


#average sales per restaurent
avg_per_rest=np.average(sales_data[:,1:],axis=1)
print("Average Sales per restaurent =",avg_per_rest)

#commulative sales per restaurant
cum_sales=np.cumsum(sales_data[:,1:],axis=1)
print("Cummulative sum per retaurant \n= ",cum_sales)
```

```python
plt.Figure(figsize=(7,6))
plt.plot(np.mean(cum_sales,axis=0))
plt.title("Average cummulative sales amoung the restaurants ")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
